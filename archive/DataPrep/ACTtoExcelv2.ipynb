```python
import pandas as pd, docx2txt
from io import StringIO
import os
import re
```

```python
#read .docx into jupyter
text = docx2txt.process ("C://Users/Bryant/Documents/GitHub/ACT/archive/DataPrep/ACT Math Sheets_v2_3_9_18/Lines.docx")
```

```python
#split text into list separated by \n\n i.e. 'enter' button
text_split = [line for line in StringIO(text).read().split('\n\n')]

#remove empty entries and spaces
text_split = list(filter(None, text_split))
```

```python
text_split
```

```python
#store topic
topic = text_split[0]
full_list = []
new = []
i = 0
for n, line in enumerate(text_split[1:]):
    #skip pic
    if re.search('picture',line.lower()):
        continue
        
    #append answers
    if i > 0:
        new.append(line)
        i -= 1
        continue
    #store month and year of test date
    if re.search('(^December|June|April|July)', line):
        month = re.search('(^December|June|April|July)', line).group()
        year = re.search('$2010|2011|2012|2013|2014|2015|2016|2017|2018|2019|2020',line).group()
        #continue to next iteration if test date is found
        continue
        
        
    #look for question number to initiate a new row entry
    if re.search('^..\.', line):
        #grab qNum
        num = re.search('^..\.', line).group().replace(".","")
        if len(new) > 0:
            full_list.append(new)
            new = []
        new.append(month[0:3] + "." + year + "." + num)
        new.append(topic)
        new.append(month + "." + year)
        new.append(num)
        #question statement
        new.append(line)
        i=5
        continue

    #same as previous block with exception re.search looking for single digits
    if re.search('^.\.', line):
        #grab qNum
        num = re.search('^.\.', line).group().replace(".","")
        if len(new) > 0:
            full_list.append(new)
            new = []
        new.append(month[0:3] + "." + year + "." + num)
        new.append(topic)
        new.append(month + "." + year)
        new.append(num)
        new.append(line)
        i=5
        continue
        
        #same as previous block with exception re.search looking for single digits
    if re.search('^..\)', line):
        #grab qNum
        num = re.search('^..\)', line).group().replace(")","")
        if len(new) > 0:
            full_list.append(new)
            new = []
        new.append(month[0:3] + "." + year + "." + num)
        new.append(topic)
        new.append(month + "." + year)
        new.append(num)
        new.append(line)
        i=5
        continue
        
    #same as previous block with exception re.search looking for single digits
    if re.search('^.\)', line):
        #grab qNum
        num = re.search('^.\)', line).group().replace(")","")
        if len(new) > 0:
            full_list.append(new)
            new = []
        new.append(month[0:3] + "." + year + "." + num)
        new.append(topic)
        new.append(month + "." + year)
        new.append(num)
        new.append(line)
        i=5
        continue
full_list.append(new)
```

```python
#prep for import into MySQL db

from pandas import DataFrame
#convert list to df
df = DataFrame.from_records(full_list)
#export to .csv excluding index and header
df.to_csv('Lines.csv', index=False, header=False)
```
